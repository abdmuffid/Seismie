```python
# ==============================================================================
# 1. PERSIAPAN DATA DAN PUSTAKA (LIBRARY)
# ==============================================================================

# Import pustaka yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Set gaya visualisasi
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# --- MULAI DARI SINI JIKA ANDA SUDAH PUNYA df_combined ---
# Jika Anda sudah memuat data dari file (misalnya CSV), Anda bisa melewati bagian pembuatan DataFrame di bawah.
# Contoh: df_combined = pd.read_csv('data_gempa.csv')
# ==============================================================================
# 2. MEMBUAT DATAFRAME CONTOH (LEWATI BAGIAN INI JIKA DATA SUDAH ADA)
# ==============================================================================
# Saya membuat DataFrame contoh berdasarkan data yang Anda berikan
# agar kode ini dapat dijalankan langsung.

df_combined


# ==============================================================================
# 3. PRE-PROCESSING DATA
# ==============================================================================

# Pilih fitur yang akan digunakan untuk clustering
# Kita menggunakan fitur numerik yang mendeskripsikan karakteristik fisik gempa
features = [ 'depth', 'magnitude']
df_features = df_combined[features].copy()

# Hapus baris yang memiliki nilai kosong (jika ada)
df_features.dropna(inplace=True)

# Scaling data sangat penting untuk K-Means karena algoritma ini berbasis jarak (distance-based)
# StandardScaler mengubah data sehingga memiliki mean=0 dan standar deviasi=1
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_features)

# ==============================================================================
# 4. MENENTUKAN JUMLAH CLUSTER (k) OPTIMAL
# ==============================================================================

# Tentukan rentang nilai k yang ingin diuji
k_range = range(2, 11)
silhouette_scores = []
db_scores = []

for k in k_range:
    # Inisialisasi dan latih model K-Means
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(scaled_features)
    
    # Hitung Silhouette Score
    # Semakin tinggi nilai (mendekati 1), semakin baik cluster-nya
    sil_score = silhouette_score(scaled_features, labels)
    silhouette_scores.append(sil_score)
    
    # Hitung Davies-Bouldin Index
    # Semakin rendah nilai (mendekati 0), semakin baik cluster-nya
    db_score = davies_bouldin_score(scaled_features, labels)
    db_scores.append(db_score)

# Visualisasi metrik untuk menentukan k optimal
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot Silhouette Score
ax1.plot(k_range, silhouette_scores, 'bo-', label='Silhouette Score')
ax1.set_xlabel('Jumlah Cluster (k)')
ax1.set_ylabel('Silhouette Score', color='b')
ax1.tick_params('y', colors='b')
ax1.set_xticks(k_range)

# Plot Davies-Bouldin Index pada sumbu y kedua
ax2 = ax1.twinx()
ax2.plot(k_range, db_scores, 'rs--', label='Davies-Bouldin Index')
ax2.set_ylabel('Davies-Bouldin Index', color='r')
ax2.tick_params('y', colors='r')

# Tambahkan judul dan legenda
plt.title('Metrik Evaluasi untuk Menentukan Jumlah Cluster (k) Optimal')
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.show()

# ==============================================================================
# 5. MELAKUKAN CLUSTERING DENGAN k OPTIMAL
# ==============================================================================

# --- Cara Membaca Grafik ---
# 1. Silhouette Score: Pilih nilai k dengan skor TERTINGGI. Puncak grafik menunjukkan cluster yang paling terpisah dan koheren.
# 2. Davies-Bouldin Index: Pilih nilai k dengan skor TERENDAH. Lembah grafik menunjukkan cluster yang paling kompak dan paling terpisah satu sama lain.
# Berdasarkan grafik di atas, kita akan memilih nilai k yang dianggap paling optimal.
# Misalnya, jika Silhouette Score puncaknya di k=4 dan Davies-Bouldin Index paling rendah di k=4,
# maka k=4 adalah pilihan yang sangat baik.

# Ganti nilai k ini berdasarkan hasil analisis grafik Anda
optimal_k = 4 

print(f"\nMelakukan clustering dengan k = {optimal_k}\n")

# Inisialisasi dan latih model K-Means final dengan k optimal
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_features['cluster'] = kmeans_final.fit_predict(scaled_features)

# Tambahkan label cluster ke DataFrame asli untuk analisis lebih lanjut
df_combined['cluster'] = df_features['cluster']

# ==============================================================================
# 6. ANALISIS DAN VISUALISASI HASIL CLUSTERING
# ==============================================================================

# Visualisasi cluster pada peta (longitude vs latitude)
plt.figure(figsize=(14, 10))
scatter = sns.scatterplot(
    data=df_combined,
    x='longitude',
    y='latitude',
    hue='cluster',
    palette='viridis',
    size='magnitude', # Ukuran titik berdasarkan magnitudo
    sizes=(20, 200),
    alpha=0.7
)

plt.title(f'Visualisasi Cluster Gempa (k={optimal_k})\nUkuran titik merepresentasikan Magnitudo', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# Analisis karakteristik setiap cluster
print("Analisis Karakteristik Rata-Rata per Cluster:")
cluster_analysis = df_combined.groupby('cluster')[features].mean()
print(cluster_analysis)
```
